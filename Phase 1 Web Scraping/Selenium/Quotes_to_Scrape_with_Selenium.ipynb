```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import csv
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
dr = webdriver.Chrome(options=options)
```

```python
# Prepare the CSV file to save the scraped data
with open('quotes.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Quote', 'Author', 'Tags'])  # Header row

    # Loop through all the pages
    while True:
        # Find all quotes on the current page
        quotes = dr.find_elements(By.CLASS_NAME, 'quote')

        for quote in quotes:
            text = quote.find_element(By.CLASS_NAME, 'text').text  # Quote text
            author = quote.find_element(By.CLASS_NAME, 'author').text  # Author name
            tags = ', '.join([tag.text for tag in quote.find_elements(By.CLASS_NAME, 'tag')])  # Tags

            # Write the data to the CSV file
            writer.writerow([text, author, tags])

        # Check if there is a next page button and click it
        try:
            next_button = dr.find_element(By.CLASS_NAME, 'next')
            next_button.find_element(By.TAG_NAME, 'a').click()
        except:
            break  # Exit the loop if no next page
```
